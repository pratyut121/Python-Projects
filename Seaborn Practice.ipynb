```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
```

```python
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=tips);
```

```python
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips);
```

```python
sns.catplot(x="size", y="total_bill", kind="swarm",
            data=tips.query("size != 4"));
```

```python
sns.catplot(x="smoker", y="tip", order=[ "Yes", "No"], data=tips,kind='swarm');
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips);
```

```python
sns.catplot(x="total_bill", y="day", kind="box", data=tips);
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="violin", data=tips);
```

```python
g = sns.catplot(x="day", y="total_bill", kind="violin", inner=None, data=tips)
sns.swarmplot(x="day", y="total_bill", color="k", size=3, data=tips, ax=g.ax);
```

```python
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind="bar",palette="ch:.50", data=titanic);
```
